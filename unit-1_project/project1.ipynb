### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# function to check if a given number is a palindrome
def is_palindrom(num):
    start = 0
    end = len(str(num)) - 1
    n = str(num)

    while start <= end:
        if n[start] != n[end]:
            return False
        start += 1
        end -= 1

    return True

# TEST: is_palindrom(num)
# output = is_palindrom(997799)
# print(output)

# function to find the largent palindrome made from the product of two three-digit numbers
def largest_palindrome():
    large_i = large_j = largest = 0

    for i in range(999, 100, -1):
        for j in range(999, 100, -1):
            num = i*j
            if is_palindrom(num):
                if num > largest:
                    largest = num
                    large_i = i
                    large_j = j
    return large_i, large_j

i, j = largest_palindrome()
print(i, j, i*j)
```

### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Function to check if the given number is a prime number. 
def is_prime(num):
    for i in range(2, num//2):
        if num%i == 0:
            return False
    return True


primes_below_2000_list = [i for i in range(2000) if is_prime(i)]
print(sum(primes_below_2000_list))
```

### Explanation

The function `is_prime` checks if a given number is a prime number. 
It uses a loop to divide the given number by all numbers starting from 2 to the half of the given input number. 
For any of this division if the modulus is 0, then it indicates the number is not a `prime` number. 

The time complexity of this function is O(n/2) => O(n).

The 2nd part of the program, uses a for loop to find the list of prime numbers between 1 to 2000 and then adds all the elements of the list to get the final result which is 277055.

### Challenge 4: String Compressor
Write a function to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def string_composer(input_string, is_case_sensitive=True):
    if not input_string:
        return ''

    if not is_case_sensitive:
        input_string = input_string.lower()

    run_length_encoded_str = []

    char_val = input_string[0]
    char_count = 1

    for ind, c in enumerate(input_string[:-1]):

        if input_string[ind+1] == char_val:
            char_count += 1
        else:
            run_length_encoded_str.append(char_val)
            run_length_encoded_str.append(str(char_count))
            char_val = input_string[ind + 1]
            char_count = 1
    run_length_encoded_str.append(char_val)
    run_length_encoded_str.append(str(char_count))

    encoded_str = ''.join(run_length_encoded_str)
    return input_string if len(encoded_str) > len(input_string) else encoded_str


print(string_composer('aabcccccaaa'))
print(string_composer('aaaaabbbb'))
print(string_composer('abcde'))
print(string_composer(''))
print(string_composer('aAbbbDDD'))
print(string_composer('aAbbbDDD', False))
print(string_composer('aaaAAAddD', False))
```

### Explanation


This solution scans the input string from the begining. 
It starts with the first character by setting a counter to 1. It increments the counter until the consequtive character elements are same. Whenever it finds a new character which is not same as the previous one, it adds the previously tracked chracter and its corresponding counter to a list and start tracking the occurence of the new character. It runs in linear time. The time complexity is O(n). 

By default, the function is case-sensitive(of input). The user of the function can override this behaviour by setting the `is_case_sensitive` parameter to `False`.

```python
from collections import namedtuple
import re


def row_conversion(row, dtypes, header_values):
    for element_ind, row_element in enumerate(row):

        try:
            column_header = header_values[element_ind]
        except Exception as e:
            print(e)
            print("element_ind: {}".format(element_ind))

        if column_header in dtypes:
            column_type = dtypes[column_header]
            if column_type == int:
                try:
                    if row_element:
                        row[element_ind] = int(row_element)
                except Exception as e:
                    print("Exception: Type Conversion failed: e:".format(e))
                    if not row_element:
                        row_element = ""
                        print("ColumnType: {}, Column_header: {}, ColumnValue: {}".format(column_type, column_header,
                                                                                          row_element))

    return row


def read_csv(filename, dtypes={}, encoding='utf-8'):
    """ Returns a list of namedtuples.
    The fields in each namedtuple are specified by the header row."""
    rows = []

    # Your code here!
    try:
        datafile = open(filename, 'r')
        header_line = datafile.readline()
    except Exception as e:
        print("Exception: {}".format(e))

    if header_line:
        header_values = header_line[:-1].split(",")

        RowNamedTuple = namedtuple("Row", header_values)

        for row in datafile:
            if row:
                row_values = row[:-1].split(",")

            if len(dtypes) == 0:
                if len(row_values) > len(header_values):
                    continue
                    #I'll address this later. Comma withing a double-quoted element
                    # print("Parsing error. Skipping reocord")
                    # print("Error row: {}".format(row_values))
                else:
                    rowTuple = RowNamedTuple._make(row_values)
            else:
                row_values_with_type = row_conversion(row_values, dtypes, header_values)
                rowTuple = RowNamedTuple._make(row_values_with_type)
            rows.append(rowTuple)

    try:
        datafile.close()
    except Exception as e:
        print("Error while closing file: {}".format(e))

    return rows


rows = read_csv('./datasets/athletes-original.csv', dtypes={})
print(rows[:3])

# For the challenge problem only
dtypes = {'age': int, 'height': int, 'weight': int,
          'gold': int, 'silver': int, 'bronze': int, 'total': int}
rows = read_csv('./datasets/athletes-unquoted.csv', dtypes=dtypes)
print('')
print(rows[:3])
```

### *BONUS* Challenge: FizzBuzz
Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz." Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def fizz_buzz(num):
    if num%3 == 0 and num%5 == 0:
        return "FizzBuzz"
    elif num%3 == 0:
        return "Fizz"
    elif num%5 == 0:
        return "Buzz"
    else:
        return ""
        
for i in range(1,101):
    print(i, fizz_buzz(i))
    
```

### Explanation

The `fizz_buzz` function first checks if the input number is divisible by 3 as well as 5. In that cast it prints "FizzBuzz". If it is only divisible by 3, then it prints "Fizz" or else if it is divisible by 5 then it prints "Buzz". In all other cases, it returns an empty string. 
The function runs in linear time.
